# Vectorisation


```python
from topolearn.simpcomplex import AlphaComplex
from topolearn.util import make_shells, plot_graph_with_data, plot_persistance_diagram
from topolearn.persistence import PersistenceImage, PersistenceLandscape

# As an example we use two nested 4-dimensional balls
X, _ = make_shells(250, dim=4, noise=0.1)
learner = AlphaComplex()
simplices = learner.fit(X)
# Plot the two first dimension 
plot_graph_with_data(simplices.graph(X), X, alpha=0.05)

# Get the persistence pairs
pairs = simplices.birth_death_pairs()
plot_persistance_diagram(pairs)
```

## Persistence landscapes

```python
# Persistence landscape calculate each homology dimension separately
from topolearn.persistence import PersistenceLandscape
plant = PersistenceLandscape()
for dim in (1,2,3):
    mat = plant.fit(pairs,dim=dim, resolution=500)
    plant.plot()
    print(plant.matrix.shape)
```

## Persistence images

```python
from topolearn.persistence import PersistenceImage

pimg = PersistenceImage(sigma=0.1, resolution=100)
pimg.fit(pairs)
pimg.plot()
# Images are stored in the images attribute
print(pimg.images.shape)
print(pimg.as_vector().shape)
```
